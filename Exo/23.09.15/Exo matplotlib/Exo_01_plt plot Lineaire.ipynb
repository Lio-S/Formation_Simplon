1. Tracer un graphe linéaire des points (x, y) où x = [0, 1, 2, 3, 4] et y = [1, 4, 3, 8, 5].

```python
# Style « pyplot »

import numpy as np
import matplotlib.pyplot as plt
# help(plt.plot)

x=[1, 3, 4, 6]
y=[2, 3, 5, 1]

# plt.plot(x, y,)
# plt.show() # affiche la figure à l'écran

# plt.plot(x, y,'r+') # Avec croix rouge
# plt.show() # affiche la figure à l'écran

# plt.plot(x, y, 'bo') # Avec cercle bleu
# plt.show() # affiche la figure à l'écran

plt.plot(x, y, 'go--', linewidth=2, markersize=12) # rond vert
plt.show() # affiche la figure à l'écran

# plt.plot(x, y, color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12) # idem go--
plt.plot(x, y, color='green', marker='.', linestyle='-',linewidth=1, markersize=20) # point et ligne
# plt.plot(x, y,'rs', label='line 2', linewidth=2, markersize=12) # 'rs' > carré rouge
plt.show() # affiche la figure à l'écran

# Style « Orienté Objet »

import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot([1, 3, 4, 6], [2, 3, 5, 1])

plt.show() # affiche la figure à l'écran

```
