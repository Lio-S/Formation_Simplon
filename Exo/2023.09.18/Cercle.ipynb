```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2*np.pi, 100)

r = np.sqrt(1.0)

x1 = r*np.cos(theta)
x2 = r*np.sin(theta)
x3 = 0.8*r*np.cos(theta)
x4 = 0.8*r*np.sin(theta)
x5 = 0.5*r*np.cos(theta)
x6 = 0.5*r*np.sin(theta)
x7 = 0.3*r*np.cos(theta)
x8 = 0.3*r*np.sin(theta)
x9 = 1.2*r*np.cos(theta)
x10 = 1.2*r*np.sin(theta)

fig, ax = plt.subplots(1)

ax.plot(x1, x2)
ax.plot(x3, x4)
ax.plot(x5, x6)
ax.plot(x7, x8)
ax.plot(x9, x10)
ax.set_aspect(1)

plt.xlim(-1.25,1.25)
plt.ylim(-1.25,1.25)

plt.grid(linestyle='--')

plt.title('Mon titre', fontsize=8)

plt.savefig("plot_circle_matplotlib_01.png", bbox_inches='tight')

plt.show()
help(plt.scatter)
```
