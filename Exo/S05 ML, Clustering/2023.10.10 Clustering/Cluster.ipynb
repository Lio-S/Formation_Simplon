# Clustering of the data

Le clustering consiste à regrouper un ensemble d'objet de telle sorte que les objets d'un même groupe (cluster) sont plus similaires entre eux qu'avec les autres groupes.

```python
from sklearn.datasets import load_digits 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
digits = load_digits()
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(10, 3))
for ax, image, label in zip(axes, digits.images, digits.target):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation="nearest")
    ax.set_title(f"Training: {label}")
```

```python
X, y = digits["data"], digits["target"]
```

```python
X.shape
```

Il y a 1797 images de 8x8 pixels.

```python
X /= X.max()
```

```python
from matplotlib import pyplot as plt 
import seaborn as sns

# Affichez quelques images
fig, axes = plt.subplots(2, 5, figsize=(10, 5))
for i, ax in enumerate(axes.ravel()):
    ax.imshow(X[i].reshape(8, 8), cmap="gray")
    ax.set_title(f"Classe : {y[i]}")
plt.show()
```

```python
model = KMeans(n_clusters=10, n_init="auto")
```

```python
model.fit(X)
```

```python
X_pred = model.predict(X)
```

```python
# Affichez quelques images
fig, axes = plt.subplots(2, 5, figsize=(10, 5))
for i, ax in enumerate(axes.ravel()):
    ax.imshow(X[i].reshape(8, 8), cmap="gray")
    ax.set_title(f"Classe : {X_pred[i]}")
plt.show()
```

```python
metrics.rand_score(y, X_pred)
```

```python
metrics.adjusted_rand_score(y, X_pred)
```
