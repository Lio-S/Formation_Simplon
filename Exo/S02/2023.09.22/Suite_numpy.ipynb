```python
# 8. Tracer une courbe polynomiale à partir des données suivantes :
# x = np.linspace(-5, 5, 100)
# y = 3 * x**3 + 2 * x**2 - 8 * x + 5

x = np.linspace(-5, 5, 100)
y = 3 * x**3 + 2 * x**2 - 8 * x + 5

import numpy as np
import matplotlib.pyplot as plt

plt.plot(x,y) 
plt.ylabel('y = 3 * x**3 + 2 * x**2 - 8 * x + 5')
plt.xlabel("x")
plt.show()
```
