```python
#Affichage de 4 figures :
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-50, 50, 40)

categories = ['A', 'B', 'C', 'D']
valeurs = [30, 25, 15, 20]

fig_total, axs = plt.subplots(nrows=2,
                              ncols=2,
                              figsize=(12, 12))
# Hyperbole                       
axs[0, 0].plot(x, 1/x)
axs[0, 0].set(title="Hyperbole",
              xlabel="axe x",
              ylabel="axe y")
# Parabole
axs[0, 1].plot(x, x**2, "-+", label="Parabole")
axs[0, 1].set(title="Parabole",
              xlabel="axe x",
              ylabel="axe y")
axs[0, 1].plot(x, 20*x + 1000,
               "-+",
               label="Droite")
axs[0, 1].legend()
#Camenbert
axs[1, 0].pie(valeurs,
              labels=categories,
              explode = [0, 0.3, 0, 0],
              colors=["Blue", "yellow", "purple", "indigo"],
              shadow=True,
              labeldistance=0.8,
              autopct = lambda x: str(round(x, 2)) + '%',
              pctdistance = 0.4)
axs[1, 0].set(title="Camembert")
# Nuage de point
axs[1, 1].scatter(np.linspace(-200, 200, 500), np.random.randn(500))
axs[1, 1].axhline(np.random.randn(500).mean(),
                  linestyle='--',
                  c="black")
axs[1, 1].set(title="Nuage de points",
              xlabel="axe x",
              ylabel="axe y")

fig_total.suptitle(t="Affichage de quatre figures", fontsize = 16, fontweight = "bold");
```
