Tracer un diagramme en secteurs (camembert) à partir des données suivantes :
categories = ['A', 'B', 'C', 'D'], valeurs = [30, 25, 15, 20]

```python
import matplotlib.pyplot as plt  # Module pour tracer les graphiques
# *****************Camembert simple :***************************
# Données d'exemple
labels = ["A", "B", "C", "D","E"]
sizes = [30, 25, 15, 20,10]
plt.figure(figsize = (5, 5),) # Taille
plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Exemple de graphique circulaire") # Ajout de titre
plt.legend()
plt.show()
# Différents paramètres :
# plt.normalize = True # : fait un camembert complet en normalisant la somme des valeurs.
# plt.colors = ['red', 'green', 'yellow'] # : séquence de couleurs à utiliser pour les parts (recyclé si pas assez de couleurs).
# plt.labeldistance = 1.2 # : la distance des labels au centre (> 1 pour être à l'extérieur du cercle).
# plt.autopct = lambda x: str(round(x, 2)) + '%' # : une fonction qui prend le pourcentage de la part et renvoie ce qui doit être affiché pour ce pourcentage.
# plt.pctdistance = 0.7 # : le distance au centre à laquelle le pourcentage précédent doit être affiché (1 = sur le cercle).
# plt.shadow = True # : indique qu'il faut afficher une ombre
# plt.title("Exemple 2") # Ajout de titre
```

```python

# On peut empêcher la normalisation en dessinant des camemberts incomplets, 
# avec une somme de valeur plus petite que 1 :

plt.figure(figsize = (8, 8))
x = [0.1, 0.2, 0.3, 0.1]
plt.pie(x, labels = ['A', 'B', 'C', 'D'], normalize = False)
plt.legend()
```

```python

# Exemple plus complet :
plt.figure(figsize = (8, 8))
x = [1, 2, 3, 4, 10]
plt.pie(x, labels = ['A', 'B', 'C', 'D', 'E'],
           colors = ['red', 'green', 'yellow'],
           explode = [0, 0.2, 0, 0, 0],
           autopct = lambda x: str(round(x, 2)) + '%',
           pctdistance = 0.7, labeldistance = 1.4,
           shadow = True)
plt.legend()
```
